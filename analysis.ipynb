```python
import jieba 
import re
import os
import pandas as pd

def is_chinese_word(word):
    for ch in word:
        if ch < '一' or ch > '龥':
            return False
    return True
def analysis(file):
    with open(file, 'r',encoding='utf-8') as f:
        txt=f.read()
        txt = re.sub("【.*?】", "", txt, count=0, flags=0)
    words=jieba.lcut(txt)
    counts={}
    for word in words:       
        counts[word]=counts.get(word,0)+1
    remove_list = []
    num = 0
    for word in counts:
        if not is_chinese_word(word):
            remove_list.append(word)       
        else:
            num += len(word) * counts[word]
    for word in remove_list:
        del counts[word]
    return counts, num
fls = os.listdir("books")
files = {}
for f in fls:
    files[f[:-4]] = os.path.join("books", f)
counts = {}
nums = {}
for f in files:
    counts[f], nums[f] = analysis(files[f])
    print(f, f"字数:{nums[f]}")   
```

```python
word_list = {}
for f in files:
    for word in counts[f]:
        if len(word) == 1:
            if word in word_list:
                word_list[word] += counts[f][word]
            else:
                word_list[word] = counts[f][word]
word_list = sorted(word_list.items(),key=lambda s: -s[1])[:50]
word_list = [x[0] for x in word_list]
for w in word_list:
    print(w, end='  ')
    
```

```python
res = {}
for word in word_list:
    res[word] = {}
    for f in counts:
        num = 0 if word not in counts[f] else counts[f][word]
        res[word][f] = num / nums[f]
res = pd.DataFrame(res).T
res
```

```python
dis = {}
for c in res.columns:
    dis[c] = {}
    for cc in res.columns:
        dis[c][cc] = int((res[c] - res[cc]).std() * 10000)
dis = pd.DataFrame(dis)
dis
        
```
